### Data Visualization:

#### _| A picture is worth a thousand words._
Now the goal is to perform univariate, bivariate and multivariate analysis to see the distribution and relationship between variables.

#### *1) Univariate Analysis:*
- The age distribution across the range of 18 to 85 exhibits a varied pattern. It typically shows a gradual increase from 18 to a peak in the 30s, followed by a gradual decline in frequency towards the higher age ranges.

```python
# Set the style and color palette
sns.set(style="white")

# Create the bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.histplot(df['Age'], bins=range(18, 86, 5), kde=True, color='#87CEEB', edgecolor='black')

# Customize the plot
bar_plot.set_title('Age Distribution', fontsize=15)
bar_plot.set_xlabel('Age', fontsize=12)
bar_plot.set_ylabel('Frequency', fontsize=12)

# Display the plot
plt.show()
```

- The dataset includes a greater proportion of females than males.

```python
# Set the style
sns.set(style="white")

# Create the bar plot for gender distribution
plt.figure(figsize=(8, 6))
bar_plot = sns.countplot(data=df, x='Gender', palette='Blues',edgecolor='black')

# Customize the plot
bar_plot.set_title('Distribution of Gender', fontsize=15)
bar_plot.set_xlabel('Gender', fontsize=12)
bar_plot.set_ylabel('Count', fontsize=12)

# Display the plot
plt.show()
```

- The most common medical conditions observed were hypertension, asthma, diabetes, obesity, and arthritis, in that order.

```python
# Count the frequency of each medical condition
condition_counts = df['Medical Condition'].value_counts()

# Set custom colors for each bar
custom_colors = sns.color_palette('pastel', len(condition_counts))

# Set the style
sns.set(style="white")

# Create the horizontal bar plot with custom colors
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x=condition_counts.values, y=condition_counts.index, palette=custom_colors)

# Customize the plot
bar_plot.set_title('Distribution of Medical Conditions', fontsize=15)
bar_plot.set_xlabel('Count', fontsize=12)
bar_plot.set_ylabel('Medical Condition', fontsize=12)

# Display the plot
plt.show()
```

- The distribution of admission types for patients shows that the highest number of admissions falls under the "Emergency" category. Followed by "Urgent" admissions, and then "Elective" admissions.

```python
# Count the frequency of each admission type
admission_counts = df['Admission Type'].value_counts()

# Set the style
sns.set(style="white")

# Create the bar plot for admission types distribution
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x=admission_counts.index, y=admission_counts.values, palette='Set2')

# Customize the plot
bar_plot.set_title('Distribution of Admission Types', fontsize=15)
bar_plot.set_xlabel('Admission Type', fontsize=12)
bar_plot.set_ylabel('Count', fontsize=12)

# Display the plot
plt.show()
```

- In terms of length of stay, the majority of patients spend 7 dayes or less than a day, indicating brief visits. However, there is proportion of patients stay for an extended period, with the maximum duration exceeding 15 days.

```python
# Set the style
sns.set(style="white")

# Create the histogram with KDE for length of stay
plt.figure(figsize=(10, 6))
hist_plot = sns.histplot(df['Length of Stay'], kde=True, bins=15, color='skyblue', edgecolor='black')

# Customize the plot
hist_plot.set_title('Distribution of Length of Stay', fontsize=15)
hist_plot.set_xlabel('Length of Stay (days)', fontsize=12)
hist_plot.set_ylabel('Frequency', fontsize=12)

# Display the plot
plt.show()
```

```python
# Count the frequency of each insurance provider
insurance_counts = df['Insurance Provider'].value_counts()

# Sort the values in descending order
insurance_counts = insurance_counts.sort_values(ascending=False)

# Set the style
sns.set(style="white")

# Create the horizontal bar plot for insurance providers distribution
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x=insurance_counts.values, y=insurance_counts.index, palette='pastel')

# Customize the plot
bar_plot.set_title('Distribution of Insurance Providers', fontsize=15)
bar_plot.set_xlabel('Count', fontsize=12)
bar_plot.set_ylabel('Insurance Provider', fontsize=12)

# Display the plot
plt.show()
```

- Test results often lack of clarity, abnormal then normal.

```python
# Count the frequency of each test result
test_results_counts = df['Test Result'].value_counts()

# Set the style
sns.set(style="white")

# Create the bar plot for test results
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x=test_results_counts.index, y=test_results_counts.values, palette='pastel')

# Customize the plot
bar_plot.set_title('Distribution of Test Results', fontsize=15)
bar_plot.set_xlabel('Test Result', fontsize=12)
bar_plot.set_ylabel('Count', fontsize=12)

# Display the plot
plt.show()
```

- The medications used by patients, listed in order of frequency, are paracetamol, penicillin, ibuprofen, Lipitor, and aspirin.

```python
# Count the frequency of each medication
medication_counts = df['Medication'].value_counts()

# Define the order of medications as listed in the problem statement
medications_order = ['Paracetamol', 'Penicillin', 'Ibuprofen', 'Lipitor', 'Aspirin']

# Subset the medication counts based on the defined order
medication_counts = medication_counts[medications_order]

# Set the style
sns.set(style="white")

# Create the horizontal bar plot for medications
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x=medication_counts.values, y=medication_counts.index, palette='pastel')

# Customize the plot
bar_plot.set_title('Distribution of Medications Used by Patients', fontsize=15)
bar_plot.set_xlabel('Count', fontsize=12)
bar_plot.set_ylabel('Medication', fontsize=12)

# Display the plot
plt.show()
```

- Distribution of Billing: Patient billing amounts range from approximately $1,200 to $8,000

```python
# Set the style
sns.set(style="white")

# Create the histogram for billing amounts
plt.figure(figsize=(10, 6))
hist_plot = sns.histplot(df['Amount Billing'], bins=8, kde=True, color='skyblue')

# Customize the plot
hist_plot.set_title('Distribution of Billing Amounts', fontsize=15)
hist_plot.set_xlabel('Billing Amount ($)', fontsize=12)
hist_plot.set_ylabel('Frequency', fontsize=12)

# Display the plot
plt.show()
```

##### **2) Bivariate Analysis:**
- The scatterplot indicates a positive correlation between age and billing amounts.

```python
# Set the style
sns.set(style="white")

# Create scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='Age', y='Amount Billing', data=df, color='skyblue')

# Customize the plot
scatter_plot.set_title('Age vs Billing Amounts', fontsize=15)
scatter_plot.set_xlabel('Age', fontsize=12)
scatter_plot.set_ylabel('Billing Amount ($)', fontsize=12)

# Display the plot
plt.show()
```

- Since females outnumber males, they contribute more to the total billing amount.

```python
# Set the style
sns.set(style="white")

# Create box plot
plt.figure(figsize=(10, 6))
box_plot = sns.boxplot(x='Gender', y='Amount Billing', data=df, palette='pastel')

# Customize the plot
box_plot.set_title('Gender vs Billing Amount', fontsize=15)
box_plot.set_xlabel('Gender', fontsize=12)
box_plot.set_ylabel('Billing Amount ($)', fontsize=12)

# Display the plot
plt.show()
```

- Box plots: These box plots display the distribution of billing amounts across different combinations of insurance providers, medical conditions, and admission types. They offer insights into the variability of billing amounts, showcasing outliers, as well as the average, maximum, and minimum values for each combination. By examining these relationships, they aid in understanding how healthcare expenses vary based on insurance coverage, medical condition, and admission type.

```python
sns.set(style="white")

# Create box plot
plt.figure(figsize=(12, 8))
box_plot = sns.boxplot(x='Medical Condition', y='Amount Billing', data=df, palette='pastel')

# Customize the plot
box_plot.set_title('Medical Conditions vs Billing Amount', fontsize=15)
box_plot.set_xlabel('Medical Condition', fontsize=12)
box_plot.set_ylabel('Billing Amount', fontsize=12)
box_plot.set_xticklabels(box_plot.get_xticklabels(), rotation=45)  # Rotate x-axis labels for better readability

# Display the plot
plt.show()

# Create box plot
plt.figure(figsize=(12, 8))
box_plot = sns.boxplot(x='Insurance Provider', y='Amount Billing', data=df, palette='pastel')

# Customize the plot
box_plot.set_title('Insurance Provider vs Billing Amount', fontsize=15)
box_plot.set_xlabel('Insurance Provider', fontsize=12)
box_plot.set_ylabel('Billing Amount', fontsize=12)
box_plot.set_xticklabels(box_plot.get_xticklabels(), rotation=45)  # Rotate x-axis labels for better readability

# Display the plot
plt.show()

# Create box plot
plt.figure(figsize=(12, 8))
box_plot = sns.boxplot(x='Admission Type', y='Amount Billing', data=df, palette='pastel')

# Customize the plot
box_plot.set_title('Admission Type vs Billing Amount', fontsize=15)
box_plot.set_xlabel('Admission Type', fontsize=12)
box_plot.set_ylabel('Billing Amount', fontsize=12)
box_plot.set_xticklabels(box_plot.get_xticklabels(), rotation=45)  # Rotate x-axis labels for better readability

# Display the plot
plt.show()
```

#### 3) Multivariate Analysis:
- **Pair plot:** is a visual tool that displays scatterplots of multiple variables against each other in a dataset, providing a quick overview of their relationships. 
- **Through the Pair plot**, it's apparent that there exists a positive correlation between Age, Billing Amount, and Length of Stay.

```python
# Selecting only the columns 'Age', 'Amount Billing', and 'Length of Stay'
selected_columns = ['Age', 'Amount Billing', 'Length of Stay']
df_selected = df[selected_columns]

# Create pair plot
sns.pairplot(df_selected)

# Display the plot
plt.show()
```

- Employing multivariate analysis with a box plot to delve into the relationship between billing amounts by admission type and Insurance Provider.

```python
# Set the style
sns.set(style="white")

# Create box plot
plt.figure(figsize=(12, 8))
box_plot = sns.boxplot(x='Admission Type', y='Amount Billing', hue='Insurance Provider', data=df, palette='Set2')

# Customize the plot
box_plot.set_title('Billing Amount Variation by Admission Type and Insurance Provider', fontsize=15)
box_plot.set_xlabel('Admission Type', fontsize=12)
box_plot.set_ylabel('Billing Amount', fontsize=12)
box_plot.legend(title='Insurance Provider', title_fontsize='13', loc='upper right')

# Display the plot
plt.show()
```

- Employing multivariate analysis with a box plot to delve into the relationship between billing amounts by gender and medical condition.

```python
# Set the style
sns.set(style="white")

# Create box plot
plt.figure(figsize=(12, 8))
box_plot = sns.boxplot(x='Gender', y='Amount Billing', hue='Medical Condition', data=df, palette='Set2')

# Customize the plot
box_plot.set_title('Billing Amounts Variation by Gender and Medical Condition', fontsize=15)
box_plot.set_xlabel('Gender', fontsize=12)
box_plot.set_ylabel('Billing Amount', fontsize=12)
box_plot.legend(title='Gender', title_fontsize='13', loc='upper right')

# Display the plot
plt.show()
```

- **Correlaation Matrix:** Based on the correlation matrix results, there appears to be a range of positive to strong positive correlations among these variables.

```python
correlation_matrix = df[['Age', 'Amount Billing', 'Length of Stay']].corr()
# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
