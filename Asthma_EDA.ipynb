# **Radar Chart for a Specific Patient**

```python
# 4. Radar chart
categories = ['Age', 'BMI', 'LungFunctionFEV1']
N = len(categories)

angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]

def radar_chart(row):
    values = df.loc[row, categories].values.flatten().tolist()
    values += values[:1]
    ax = plt.subplot(111, polar=True)
    ax.fill(angles, values, color='blue', alpha=0.25)
    ax.plot(angles, values, color='blue', linewidth=2)
    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)
    plt.title(f'Patient {row}')
    plt.show()

# Example radar chart for the first patient
radar_chart(0)

# 5. Facet Grid
g = sns.FacetGrid(df, col="Diagnosis", row="Gender")
g.map(sns.histplot, "Age")
plt.show()
```
